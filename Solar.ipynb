# **400 MW Solar and Wind Hybrid Power Plant in Surat**

This notebook provides a detailed pre-feasibility analysis for setting up a 400 MW solar and wind hybrid power plant on the outskirts of Surat, Gujarat. The analysis includes market potential, technical feasibility, financial projections, and risk assessment, with visualizations using Matplotlib and data analysis using Scikit-learn.


```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')
```

## Data Preparation

We'll start by loading and preparing the data for analysis. This includes data on solar and wind potential, costs, and revenue projections.


```python
import pandas as pd
data = {
    'Resource': ['Solar', 'Wind'],
    'Average Irradiance (kWh/m²/day)': [5.5, None],  # Placeholder for solar data
    'Average Wind Speed (m/s)': [None, 7.0],  # Placeholder for wind data
    'Annual Production (MWh)': [2000000, 1000000],  # Hypothetical data
    'Cost (₹ Crore)': [150, 200],  # Hypothetical cost data
    'Revenue (₹ Crore)': [90, 80]  # Hypothetical revenue data
}

df = pd.DataFrame(data)
df
```

## Market Analysis

We analyze the market potential for solar and wind energy in Surat, including demand trends and competitive landscape.


```python
fig, ax = plt.subplots(figsize=(10, 6))
df.plot(kind='bar', x='Resource', y=['Cost (₹ Crore)', 'Revenue (₹ Crore)'], ax=ax)
ax.set_title('Cost and Revenue Comparison for Solar and Wind Energy')
ax.set_ylabel('Amount (₹ Crore)')
plt.xticks(rotation=0)
plt.show()
```

## Technical Feasibility

This section evaluates the technical aspects of the project, including resource availability, technology, and potential challenges.


```python
# Visualization of average production by resource
fig, ax = plt.subplots(figsize=(10, 6))
df.plot(kind='pie', y='Annual Production (MWh)', labels=df['Resource'], autopct='%1.1f%%', ax=ax)
ax.set_title('Annual Production by Resource')
plt.show()
```

## Financial Projections

Here, we calculate the financial metrics for the project, including capital investment, operating costs, and revenue projections.


```python
# Financial metrics calculation
initial_investment = 350  # ₹ Crore
annual_revenue = df['Revenue (₹ Crore)'].sum()
annual_cost = df['Cost (₹ Crore)'].sum()

roi = (annual_revenue - annual_cost) / initial_investment * 100  # Return on Investment in percentage
break_even_point = initial_investment / (annual_revenue - annual_cost)

roi, break_even_point
```

## Risk Analysis

This section identifies the major risks associated with the project and suggests mitigation strategies.


```python
# Simple risk analysis visualization
risks = {
    'Regulatory Changes': 30,
    'Technical Challenges': 25,
    'Financial Risks': 20,
    'Operational Issues': 15,
    'Market Risks': 10
}

fig, ax = plt.subplots(figsize=(10, 6))
plt.pie(risks.values(), labels=risks.keys(), autopct='%1.1f%%', startangle=140)
ax.set_title('Risk Distribution')
plt.show()
```
